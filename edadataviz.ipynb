First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
# Plot a scatter point chart with x axis to be Flight Number and y axis to be the launch site, and hue to be the class value
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(
    x='FlightNumber',
    y='LaunchSite',
    hue='Class',
    data=df,
    kind='strip',
    height=6,
    aspect=2
)

plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.title("Flight Number vs Launch Site (Colored by Class)")
plt.show()
```

Now try to explain the patterns you found in the Flight Number vs. Launch Site scatter point plots.


```python
# Patterns observed in the Flight Number vs Launch Site scatter plot:
#
# 1. Overall success rate increases with flight number.
#    - Early flights (low FlightNumber) show many failures (Class = 0).
#    - As FlightNumber increases, most points become successful (Class = 1).
#    - This reflects SpaceX improving landing reliability over time.
#
# 2. Launch sites show different usage patterns:
#
#    CCAFS SLC-40:
#    - Has the largest number of launches.
#    - Early flights include many failures.
#    - Later flights become mostly successful, showing a clear learning curve.
#
#    VAFB SLC-4E:
#    - Fewer total launches.
#    - Early launches show a mix of failures and successes.
#    - Later flights trend toward more consistent success.
#
#    KSC LC-39A:
#    - Begins appearing only at mid-to-high flight numbers.
#    - Almost all launches here are successful.
#    - Indicates SpaceX began using this site after achieving more reliable landings.
#
# 3. Newer launch sites tend to start with higher success rates.
#    - LC-39A launches start later and show immediate high reliability.
#
# 4. General conclusion:
#    - SpaceX shows a clear improvement in landing success over time.
#    - Different launch sites have different roles and start times, but all follow the trend of increasing success.
```

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
import matplotlib.pyplot as plt

# Compute success rate for each orbit (mean of Class)
orbit_success = df.groupby('Orbit')['Class'].mean()

# Plot bar chart
orbit_success.plot(kind='bar', figsize=(10,6))

plt.xlabel('Orbit Type')
plt.ylabel('Success Rate')
plt.title('Success Rate by Orbit Type')
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.


```python
# Analysis of Success Rate by Orbit Type:
#
# - ES-L1, GEO, HEO, and SSO have a 100% success rate.
#   These orbit types show perfect performance with no recorded failures.
#
# - VLEO also has a very high success rate (around 0.85), indicating strong reliability.
#
# - LEO, MEO, PO, and ISS show moderate success rates (around 0.65–0.72).
#   These orbits have some failures but are generally reliable.
#
# - GTO has the lowest success rate (around 0.52).
#   GTO missions appear to be more challenging and show more landing failures.
#
# Summary:
# The highest success rates occur for ES-L1, GEO, HEO, and SSO orbits.
# GTO has the lowest success rate, while LEO/MEO/ISS/PO fall in the middle range.
```

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Plot a scatter point chart with x axis to be FlightNumber and y axis to be the Orbit, and hue to be the class value
# Scatter plot of FlightNumber vs Orbit, colored by Class

sns.catplot(
    data=df,
    x="FlightNumber",
    y="Orbit",
    hue="Class",
    kind="strip",
    height=6,
    aspect=2
)

plt.title("Flight Number vs Orbit (Colored by Class)")
plt.xlabel("Flight Number")
plt.ylabel("Orbit")
plt.show()
```

You can observe that in the LEO orbit, success seems to be related to the number of flights. Conversely, in the GTO orbit, there appears to be no relationship between flight number and success.


```python
# Interpretation of the FlightNumber vs Orbit scatter plot:
# 
# • LEO (Low Earth Orbit) shows a clear pattern where higher Flight Numbers
#   are associated with higher success rates, suggesting experience and
#   iterative improvements contributed to reliability.
#
# • GTO (Geostationary Transfer Orbit) does NOT show a clear relationship
#   between Flight Number and success. Success and failure are more mixed,
#   indicating that mission difficulty may play a larger role than flight order.
#
# • ISS and PO orbits show moderately improving success over time, but the
#   pattern is not as strong as in LEO.
#
# • Some orbits (like ES-L1, SO, and GEO) have very few launches, so no clear
#   trend can be observed due to limited data.
#
# Overall conclusion:
# Experience over time (higher flight numbers) improves success particularly
# for the more common LEO missions, but for more challenging or less frequent
# orbits, success does not strongly correlate with flight number.
```

### TASK  5: Visualize the relationship between Payload Mass and Orbit type


Similarly, we can plot the Payload Mass vs. Orbit scatter point charts to reveal the relationship between Payload Mass and Orbit type


```python
# Plot a scatter point chart with x axis to be Payload Mass and y axis to be the Orbit, and hue to be the class value
# Scatter plot of Payload Mass vs Orbit, colored by Class
plt.figure(figsize=(12,6))
sns.scatterplot(
    data=df,
    x='PayloadMass',
    y='Orbit',
    hue='Class',
    palette='deep'
)

plt.title("Payload Mass vs Orbit (Colored by Class)")
plt.xlabel("Payload Mass (kg)")
plt.ylabel("Orbit")
plt.legend(title='Class')
plt.show()
```

With heavy payloads the successful landing or positive landing rate are more for Polar,LEO and ISS.

However, for GTO, it's difficult to distinguish between successful and unsuccessful landings as both outcomes are present.


```python
# Interpretation of the Payload Mass vs Orbit scatter plot:
#
# • For Polar (PO), LEO, and ISS orbits, heavier payloads tend to show higher
#   success rates, suggesting these mission types handle heavy loads reliably.
#
# • GTO (Geostationary Transfer Orbit) does not show a clear distinction between
#   success and failure. Both successful and unsuccessful launches occur across
#   similar payload mass ranges, indicating that mission complexity—not mass—
#   is likely the limiting factor.
#
# • Some orbits have very few launches, making it hard to identify a strong
#   relationship between payload mass and success for those cases.
#
# Overall conclusion:
# Payload mass influences success for several orbits (LEO, ISS, PO), but for
# more challenging or less frequently flown orbits like GTO, payload mass alone
# is not a strong predictor of mission outcome.
```

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
# Group by the extracted year and compute the average success rate
yearly_success = df.groupby('Date')['Class'].mean().reset_index()

# Plot the line chart
plt.figure(figsize=(12,6))
sns.lineplot(data=yearly_success, x='Date', y='Class', marker='o')

plt.title("Yearly Launch Success Trend")
plt.xlabel("Year")
plt.ylabel("Average Success Rate")
plt.grid(True)
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020


```python
# Interpretation:
# • The success rate shows a clear upward trend beginning around 2013.
# • Prior to 2013, SpaceX missions had mixed outcomes with lower reliability.
# • From 2013 onward, the success rate steadily increases year after year,
#   reflecting major improvements in Falcon 9 launch reliability.
# • By 2018–2020, the success rate stabilizes at a very high level, indicating
#   mature rocket design, better reusability, and operational consistency.
```
